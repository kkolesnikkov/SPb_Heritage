```python
import pandas as pd
import numpy as np
import re
from collections import Counter
from pymorphy3 import MorphAnalyzer

morph = MorphAnalyzer()
```

```python
df = pd.read_csv('2_func_markers.csv', sep=',')
del df['district_norm']
df.head(10)
```

```python
func_list = [
    'могила',
    'жилой',
    'сад',
    'ограда',
    'мост',
    'статуя',
    'особняк',
    'церковь',
    'дача',
    'завод',
    'парк',
    'ворота',
    'бюст',
    'доходный',
    'фонтан'
]
```

```python
marks = []
districts = []

for distr, mark, name in zip(df['district'], df['func_markers'], df['name_object']):
    for w in name.split(' '):
        if len(w) > 2:
            w = morph.parse(w)[0].normal_form
            if w in func_list:
                districts.append(distr)
                marks.append(w)

data = {
    'district': districts,
    'func_markers': marks
}

df_d_m = pd.DataFrame(data)
df_d_m.head()
```

```python
df_d_m.to_csv('markers_to_districts.csv', index=False, encoding='utf-8-sig')
```
